# Seperating the name as individual column

```python
firstlast = pd.DataFrame({"String":["Sachin Mishra","Pawan Verma"]})
firstlast["First_Name"] = firstlast["String"].str.split(" ",expand=True)[0]
firstlast["Last_Name"] = firstlast["String"].str.split(" ", expand = True)[1]
firstlast
```

```python
firstlast = pd.DataFrame({"String":["Sachin Mishra","Pawan Verma"]})
firstlast["upper"] = firstlast["String"].str.upper()
firstlast["lower"] = firstlast["String"].str.lower()
firstlast["title"] = firstlast["String"].str.title()
firstlast
```

# Joining operations in pandas 

```python
df1 = pd.DataFrame({"Key":["A","B","C","D"], "Value": np.random.randn(4)})
df1
```

```python
df2 = pd.DataFrame({"Key":["B","D","D","E"], "Value":np.random.randn(4)})
df2
```

```python
#inner joins
inner_join = df1.merge(df2, on = ["Key"], how = "inner")
inner_join
```

```python
#left joins
left_joins = df1.merge(df2, on = ["Key"],how = "left")
left_joins
```

```python
#right joins
right_joins = df1.merge(df2, on = ["Key"],how = "right")
right_joins
```

```python
#Outer joins
outer_join = df1.merge(df2, on=["Key"], how = 'outer')
outer_join
```
